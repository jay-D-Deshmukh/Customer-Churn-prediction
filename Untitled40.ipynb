# Data Preprocessing and Model Building

```python
df
```

**Droping the columns "Name" and "CustomerID"**


```python
df.drop(columns = ['CustomerID','Name'] , inplace=True)
```

```python
df
```

```python
def encode(data):
  if data == "Male":
    return 1
  else:
    return 0
```

```python
df['Gender'] = df['Gender'].apply(encode)
```

```python
df['Location']
```

```python
df = pd.get_dummies(df, columns=['Location'],drop_first=True)
```

```python
df
```

```python
x = df[['Age','Gender','Subscription_Length_Months','Monthly_Bill','Total_Usage_GB','Location_Houston','Location_Los Angeles','Location_Miami','Location_New York']]
y = df['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Initialize and train the logistic regression model
logistic_reg = LogisticRegression(max_iter=1000)
logistic_reg.fit(x_train, y_train)

# Predict using the test set
y_pred = logistic_reg.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Generate a classification report
report = classification_report(y_test, y_pred)

# Generate a confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("\nClassification Report:")
print(report)
print("\nConfusion Matrix:")
print(confusion_mat)
```

```python
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'max_iter': [100, 500, 1000]
}

# Initialize the logistic regression model
logistic_reg = LogisticRegression()

# Create a GridSearchCV object for hyperparameter tuning
grid_search = GridSearchCV(logistic_reg, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)

# Fit the GridSearchCV to the training data
grid_search.fit(x_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Get the best estimator (model with the best hyperparameters)
best_model = grid_search.best_estimator_

# Predict using the best model
y_pred = best_model.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Generate a classification report
report = classification_report(y_test, y_pred)

# Generate a confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)

print("Best Hyperparameters:", best_params)
print(f"Accuracy: {accuracy}")
print("\nClassification Report:")
print(report)
print("\nConfusion Matrix:")
print(confusion_mat)
```
